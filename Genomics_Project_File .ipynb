# Visualizing Raw Data

Lets check the distribution of variants along the length of our species shortest chromosome (21) 

```python
#create function to examine raw variant density at different bp locations in our chromosome
def plot_windowed_variant_density(pos, window_size, title=None):
    
    # setup windows
    bins = np.arange(0, pos.max(), window_size)
    
    # use window midpoints as x coordinate
    x = (bins[1:] + bins[:-1])/2
    
    # compute variant density in each window
    h, _ = np.histogram(pos, bins=bins)
    y = h / window_size
    
    # plot
    fig, ax = plt.subplots(figsize=(12, 3))
    sns.despine(ax=ax, offset=10)
    ax.plot(x, y)
    ax.set_xlabel('Chromosome position (bp)')
    ax.set_ylabel('Variant density (bp$^{-1}$)')
    if title:
        ax.set_title(title)
```

```python
#run the function to determine our plot_windowed_variant_density
plot_windowed_variant_density(df.POS, window_size = 100000, title = 'Raw Variant Density for Chromosome 21')
```

__Variant Density along the Chromosome__

- This chart gives us an idea of what parts of the genome are more variable than others. It seems the initial 5 million bps have generally been conserved between populations, while starting around 13 million bps along Chr21 there seems to be larger amounts of variabilty
- We would expect this chart to differ greatly among all our chromosomes, however it will always be of interest to anyone studying the human genome to identify the variability of a region in our genome. 

```python
#define a function that lets us examine the variant DP or depth of each varaint read at a specific locus
def plot_variant_hist(f, bins):
    x = f
    title = 'Variant DP Distribution'
    xlabel = 'Read Depth'
    fig, ax = plt.subplots(figsize=(10, 5))
    sns.despine(ax=ax, offset=10)
    ax.hist(x, bins=bins)
    ax.set_xlabel(xlabel)
    ax.set_ylabel('No. variants')
    ax.set_title(title)
```

```python
#gives us a distribution of the read depth for all of our variant samples
plot_variant_hist(df.DP, 25)
```

__Read Depth__

- Read depth tells us how many calls our genomic sequencing were able to capture for a certain variant, essentially how many times did we discover a variant at this specific location. While not a perfect correlation this can often tell us how much we should trust the variant read at a specific loci. 
- The majority of our samples have a limited read depth for our calls on chromsome 21, this however is not entirely surprising as this is the smallest chromosome and therefore the least suited for extensive calls at certain locations. In addition we should remember that we had 2548 individuals sampled so while read depth can be higher than our NS we would not expect it to be too many times larger
- Going forward we will examine other metrics that we can use to evaluate the read quality of variants in our dataset
